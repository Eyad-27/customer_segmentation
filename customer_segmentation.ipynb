```python
# import libraries and load dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
# select feature columns and show summary
x_col = 'Annual Income (k$)'
y_col = 'Spending Score (1-100)'
df[[x_col, y_col]].describe()
```

```python
# plot scatter and histograms for the two features
plt.figure(figsize=(5,5))
plt.scatter(df[x_col], df[y_col], s=30)
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title('Income vs Spending')
plt.tight_layout()
plt.show()
fig, ax = plt.subplots(1,2, figsize=(9,3))
sns.histplot(df[x_col], bins=15, ax=ax[0])
sns.histplot(df[y_col], bins=15, ax=ax[1])
ax[0].set_title(x_col)
ax[1].set_title(y_col)
plt.tight_layout()
plt.show()
```

```python
# scale features
scaler = StandardScaler()
X = df[[x_col, y_col]].values
X_scaled = scaler.fit_transform(X)
X_scaled[:5]
```

```python
# find best k using elbow and silhouette methods
ks = range(2,11)
inertia = []
silh = []
models = {}
for k in ks:
    km = KMeans(n_clusters=k, random_state=0, n_init=10)
    labels = km.fit_predict(X_scaled)
    inertia.append(km.inertia_)
    silh.append(silhouette_score(X_scaled, labels))
    models[k] = (km, labels)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(list(ks), inertia, marker='o')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('Elbow')
plt.subplot(1,2,2)
plt.plot(list(ks), silh, marker='o', color='green')
plt.xlabel('k')
plt.ylabel('Silhouette')
plt.title('Silhouette')
plt.tight_layout()
plt.show()
best_k = int(ks[np.argmax(silh)])
best_k
```

```python
# fit best k and show clusters
km, labels = models[best_k]
df['cluster'] = labels
cent_scaled = km.cluster_centers_
cent_orig = scaler.inverse_transform(cent_scaled)
centers = pd.DataFrame(cent_orig, columns=[x_col, y_col])
centers['cluster'] = range(best_k)
centers
```

```python
# plot clusters in original space
plt.figure(figsize=(6,6))
sns.scatterplot(data=df, x=x_col, y=y_col, hue='cluster', palette='tab10', s=50)
plt.scatter(cent_orig[:,0], cent_orig[:,1], c='black', s=120, marker='X')
plt.title(f'Clusters k={best_k}')
plt.tight_layout()
plt.show()
```

```python
# plot clusters in scaled space
plt.figure(figsize=(6,6))
sns.scatterplot(x=X_scaled[:,0], y=X_scaled[:,1], hue=df['cluster'], palette='tab10', s=50)
plt.scatter(cent_scaled[:,0], cent_scaled[:,1], c='black', s=120, marker='X')
plt.title('Scaled Space')
plt.tight_layout()
plt.show()
```

```python
# show cluster sizes
df['cluster'].value_counts().sort_index()
```

```python
# try dbscan clustering
from sklearn.cluster import DBSCAN
eps_values = [0.3, 0.4, 0.5, 0.6, 0.7]
best_eps = None
best_clusters = None
for eps in eps_values:
    db = DBSCAN(eps=eps, min_samples=5)
    db_labels = db.fit_predict(X_scaled)
    n_clusters = len(set([l for l in db_labels if l != -1]))
    if n_clusters >= 2 and n_clusters <= 10:
        best_eps = eps
        best_clusters = db_labels
        break
if best_eps is None:
    best_eps = eps_values[2]
    db = DBSCAN(eps=best_eps, min_samples=5)
    best_clusters = db.fit_predict(X_scaled)
df['dbscan_cluster'] = best_clusters
unique_dbscan = sorted(set(best_clusters))
print('DBSCAN eps:', best_eps, 'clusters (excluding noise):', len([c for c in unique_dbscan if c != -1]))
plt.figure(figsize=(6,6))
palette = sns.color_palette('tab10', n_colors=max(1, len([c for c in unique_dbscan if c != -1])))
colors = []
color_map = {}
cluster_indices = [c for c in unique_dbscan if c != -1]
for i, c in enumerate(cluster_indices):
    color_map[c] = palette[i % len(palette)]
for lab in best_clusters:
    if lab == -1:
        colors.append('#999999')
    else:
        colors.append(color_map[lab])
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=colors, s=40)
plt.title('DBSCAN Clusters (scaled)')
plt.tight_layout()
plt.show()
df['dbscan_cluster'].value_counts().sort_index()
```
